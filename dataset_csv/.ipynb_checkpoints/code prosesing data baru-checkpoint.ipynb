```python
import os
import pandas as pd
```

```python
covid_path = 'ml/CT_COVID'
non_covid_path = 'ml/CT_NonCOVID'

covid = [os.path.join(covid_path, file) for file in os.listdir(covid_path)]
non_covid = [os.path.join(non_covid_path, file) for file in os.listdir(non_covid_path)]
```

```python
data = pd.DataFrame(columns=['image_path', 'label'])
```

```python
data['image_path'] = covid + non_covid
data['label'] = ['covid'] * len(covid) + ['non_covid'] * len(non_covid)
```

```python
data = data.sample(frac=1).reset_index(drop=True)
```

```python
data.head()
```

```python
data['image_path'] = data['image_path'].str.replace('ml/CT_NonCOVID/', '')
data['image_path'] = data['image_path'].str.replace('ml/CT_COVID/', '')

# Change the label name "non_covid" to a different label
data['label'] = data['label'].replace('non_covid', 'No Finding')
data['label'] = data['label'].replace('covid', 'Pneumonia/Viral/COVID-19')
```

```python
# Convert float values to strings
data2['filename'] = data2['filename'].astype(str)

# Modify values to ensure ".png" extension
data2['filename'] = data2['filename'].apply(lambda x: x if x.endswith('.png') else x + '.png')
```

```python
data = data.rename(columns={'image_path': 'filename', 'label': 'finding'})
```

```python
data3 = pd.merge(data, data2, on='filename', how='inner')
```

```python
columns_to_drop = [2, 3]  # Misalnya ingin menghapus kolom dengan indeks 1 dan 3
data3 = data3.drop(data3.columns[columns_to_drop], axis=1)
```

```python
data4 = data[(data['finding'] == 'No Finding')]
data4 = data4.reset_index(drop=True)
data4
```

```python
data_normal_cli = pd.read_csv('raw clinical.csv')
data_normal_cli = data_normal_cli['cxr_findings']
data_normal_cli = data_normal_cli.dropna(axis=0).reset_index(drop=True)
data_normal_cli = data_normal_cli.to_frame()
data_normal_cli = data_normal_cli.rename(columns={'cxr_findings': 'clinical_notes'})
data_normal_cli
```

```python
data5 = data4.join(data_normal_cli)
data5 = data5.dropna(axis=0).reset_index(drop=True)
data5
```

```python
data_combine = pd.concat([data3, data5], ignore_index=True)
data_combine
```
